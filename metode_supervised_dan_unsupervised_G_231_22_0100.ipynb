```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Load dataset Iris
iris = load_iris()
X = iris.data
y = iris.target
# Bagi dataset menjadi data latih dan data uji
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Inisialisasi model K-Nearest Neighbors
knn = KNeighborsClassifier(n_neighbors=3)

# Latih model dengan data latih
knn.fit(X_train, y_train)

# Prediksi kelas untuk data uji
y_pred = knn.predict(X_test)
# Hitung akurasi prediksi
accuracy = accuracy_score(y_test, y_pred)
print(f'Akurasi K-Nearest Neighbors: {accuracy:.2f}')
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# Membuat dataset acak dengan 500 sampel, 4 pusat, dan deviasi standar 1.0
X, _ = make_blobs(n_samples=500, centers=4, cluster_std=1.0, random_state=42)

# Inisialisasi model K-Means dengan 4 klaster
kmeans = KMeans(n_clusters=4, random_state=42)

# Melakukan klasterisasi terhadap data
kmeans.fit(X)
# Mendapatkan label klaster untuk setiap data
y_pred = kmeans.predict(X)

# Plot hasil klasterisasi
plt.figure(figsize=(8, 6))
plt.scatter(X[:, 0], X[:, 1], c=y_pred, cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], marker='o', s=200, color='red', alpha=0.9, label='Centroids')
plt.title('K-Means Clustering')
plt.legend()
plt.show()
```
